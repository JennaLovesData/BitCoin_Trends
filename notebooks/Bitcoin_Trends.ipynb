```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('../data/bitcoin_trends.csv', parse_dates = ['Timestamp'])
df
```

## Avg. over Bitcoin lifespany

```python
mean_close = df['Close'].mean()
mean_close
```

## Rolling avg. every 5 minutes

```python
df['rolling_mean'] = df['Close'].rolling(window=5).mean()
df
```

## Exponential moving avg. most recent prices than others 

```python
df['EMA_5min'] = df['Close'].ewm(span=5).mean()
df
```

```python
df['open_time'] = pd.to_datetime(df['Open'], unit='s')
df
```

```python
df['close_time'] = pd.to_datetime(df['Close'], unit='s')
df
```

## Convert Unix time to readable date format, to extract the year, month, week, and daily trends. Grouped data to prepare for analysis
*Time Based Aggergation*

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s') 
df
```

```python
df.set_index('Timestamp', inplace=True)
df
```

```python
minute_data = df
```

### Hourly Data

```python
hourly_data = df.resample('H').agg({'Open': 'first','High': 'max', 'Low': 'min', 'Close': 'last','Volume': 'sum'})
hourly_data_reset = hourly_data.reset_index()
hourly_data_reset
```

### Daily Data

```python
daily_data = hourly_data.resample('D').agg({'Open': 'first','High': 'max', 'Low': 'min', 'Close': 'last','Volume': 'sum'})
daily_data_reset = daily_data.reset_index()
daily_data_reset
```

### Monthly Data

```python
monthly_data = daily_data.resample('M').agg({'Open': 'first','High': 'max', 'Low': 'min', 'Close': 'last','Volume': 'sum'})
monthly_data_reset = monthly_data.reset_index()
monthly_data_reset
```

### Yearly Table

```python
yearly_data = monthly_data.resample('Y').agg({'Open': 'first','High': 'max', 'Low': 'min', 'Close': 'last','Volume': 'sum'})
yearly_data_reset = yearly_data.reset_index()
yearly_data_reset
```

### Weekly Table

```python
weekly_data = daily_data.resample('W').agg({'Open': 'first','High': 'max', 'Low': 'min', 'Close': 'last','Volume': 'sum'})
weekly_data_reset = weekly_data.reset_index()
weekly_data_reset
```

## Moving Averages & Indicators 
Simple and Exponential Moving Averages 

```python
hourly_data['SMA_10'] = hourly_data['Close'].rolling(window=10).mean()
hourly_data['EMA_10'] = hourly_data['Close'].ewm(span=10, adjust=False).mean()
daily_data['SMA_10'] = daily_data['Close'].rolling(window=10).mean()
daily_data['EMA_10'] = daily_data['Close'].ewm(span=10, adjust=False).mean()
weekly_data['SMA_10'] = weekly_data['Close'].rolling(window=10).mean()
weekly_data['EMA_10'] = weekly_data['Close'].ewm(span=10, adjust=False).mean()
monthly_data['SMA_10'] = monthly_data['Close'].rolling(window=10).mean()
monthly_data['EMA_10'] = monthly_data['Close'].ewm(span=10, adjust=False).mean()
yearly_data['SMA_10'] = yearly_data['Close'].rolling(window=10).mean()
yearly_data['EMA_10'] = yearly_data['Close'].ewm(span=10, adjust=False).mean()
```

## Bollinger Bands

```python
hourly_data['SMA_20'] = hourly_data['Close'].rolling(window=20).mean()
hourly_data['stddev'] = hourly_data['Close'].rolling(window=20).std()
hourly_data['Lower_BB'] = hourly_data['SMA_20'] - (hourly_data['stddev'] * 2)
hourly_data['Upper_BB'] = hourly_data['SMA_20'] + (hourly_data['stddev'] * 2)
```

```python
daily_data['SMA_20'] = daily_data['Close'].rolling(window=20).mean()
daily_data['stddev'] = daily_data['Close'].rolling(window=20).std()
daily_data['Lower_BB'] = daily_data['SMA_20'] - (daily_data['stddev'] * 2)
daily_data['Upper_BB'] = daily_data['SMA_20'] + (daily_data['stddev'] * 2)
```

```python
weekly_data['SMA_20'] = weekly_data['Close'].rolling(window=20).mean()
weekly_data['stddev'] = weekly_data['Close'].rolling(window=20).std()
weekly_data['Lower_BB'] = weekly_data['SMA_20'] - (weekly_data['stddev'] * 2)
weekly_data['Upper_BB'] = weekly_data['SMA_20'] + (weekly_data['stddev'] * 2)
```

```python
monthly_data['SMA_20'] = monthly_data['Close'].rolling(window=20).mean()
monthly_data['stddev'] = monthly_data['Close'].rolling(window=20).std()
monthly_data['Lower_BB'] = monthly_data['SMA_20'] - (monthly_data['stddev'] * 2)
monthly_data['Upper_BB'] = monthly_data['SMA_20'] + (monthly_data['stddev'] * 2)
```

## Volatility Analysis 
*Percentage Change*

```python
hourly_data['Pct_Change'] = hourly_data['Close'].pct_change() * 100
daily_data['Pct_Change'] = daily_data['Close'].pct_change() * 100
weekly_data['Pct_Change'] = weekly_data['Close'].pct_change() * 100
monthly_data['Pct_Change'] = monthly_data['Close'].pct_change() * 100
yearly_data['Pct_Change'] = yearly_data['Close'].pct_change() * 100
```

### Log Returns 
*Measures percent change in price*

```python
hourly_data['Log_Return'] = np.log(hourly_data['Close']/hourly_data['Close'].shift(1)) * 100 
daily_data['Log_Return'] = np.log(daily_data['Close']/daily_data['Close'].shift(1)) * 100
weekly_data['Log_Return'] = np.log(weekly_data['Close']/weekly_data['Close'].shift(1)) * 100
monthly_data['Log_Return'] = np.log(monthly_data['Close']/monthly_data['Close'].shift(1)) * 100
yearly_data['Log_Return'] = np.log(yearly_data['Close']/yearly_data['Close'].shift(1)) * 100
```

## Volume Analysis 
*Rolling Mean of Volume: how many people are selling/buying*

```python
hourly_data['Volume_MA'] = hourly_data['Volume'].rolling(window=10).mean()
daily_data['Volume_MA'] = daily_data['Volume'].rolling(window=10).mean()
weekly_data['Volume_MA'] = weekly_data['Volume'].rolling(window=10).mean()
monthly_data['Volume_MA'] = monthly_data['Volume'].rolling(window=10).mean()
```

### Detecting High Volume Spikes 

```python
hourly_data['High_Volume'] = hourly_data['Volume'] > hourly_data['Volume'].rolling(window=20).mean() * 2
daily_data['High_Volume'] = daily_data['Volume'] > daily_data['Volume'].rolling(window=20).mean() * 2
weekly_data['High_Volume'] = weekly_data['Volume'] > weekly_data['Volume'].rolling(window=20).mean() * 2
monthly_data['High_Volume'] = monthly_data['Volume'] > monthly_data['Volume'].rolling(window=20).mean() * 2
```

```python
high_vol_trades_hourly = hourly_data[hourly_data['High_Volume']]
high_vol_trades_daily = daily_data[daily_data['High_Volume']]
high_vol_trades_weekly = weekly_data[weekly_data['High_Volume']]
high_vol_trades_monthly = monthly_data[monthly_data['High_Volume']]
```

## Trend Analysis 
*Looks at Uptrends and Downtrends*

```python
hourly_data['Uptrend'] = hourly_data['Close'] > hourly_data['Close'].shift(1)
hourly_data['Downtrend'] = hourly_data['Close'] < hourly_data['Close'].shift(1)
daily_data['Uptrend'] = daily_data['Close'] > daily_data['Close'].shift(1)
daily_data['Downtrend'] = daily_data['Close'] < daily_data['Close'].shift(1)
weekly_data['Uptrend'] = weekly_data['Close'] > weekly_data['Close'].shift(1)
weekly_data['Downtrend'] = weekly_data['Close'] < weekly_data['Close'].shift(1)
monthly_data['Uptrend'] = monthly_data['Close'] > monthly_data['Close'].shift(1)
monthly_data['Downtrend'] = monthly_data['Close'] < monthly_data['Close'].shift(1)
yearly_data['Uptrend'] = yearly_data['Close'] > yearly_data['Close'].shift(1)
yearly_data['Downtrend'] = yearly_data['Close'] < yearly_data['Close'].shift(1)
```

Detecting Breakouts

```python
minute_data['Breakout'] = minute_data['High'] > minute_data ['High'].rolling(window=50).max()
hourly_data['Breakout'] = hourly_data['High'] > hourly_data ['High'].rolling(window=50).max()
daily_data['Breakout'] = daily_data['High'] > daily_data ['High'].rolling(window=50).max()
weekly_data['Breakout'] = weekly_data['High'] > weekly_data ['High'].rolling(window=50).max()
monthly_data['Breakout'] = monthly_data['High'] > monthly_data ['High'].rolling(window=50).max()
```
